```python
from autoscraper import AutoScraper
```

```python
amazon_url="https://www.amazon.in/s?k=iphone"
wanted_list=["₹71,990","Apple iPhone 13 (128GB) - Midnight","3,385"]
```

```python
scraper=AutoScraper()
result=scraper.build(amazon_url,wanted_list)
print(result)
```

```python
scraper.get_result_similar(amazon_url,grouped=True)#3 main info. thats why group 
```

```python
scraper.set_rule_aliases({'rule_cmg9':'Title','rule_2sei':'Price'})
scraper.keep_rules(['rule_cmg9','rule_2sei'])
scraper.save('amazon-search')
```

```python
results=scraper.get_result_similar('https://www.amazon.in/s?k=mi+phones+under+15000',group_by_alias=True)
```
